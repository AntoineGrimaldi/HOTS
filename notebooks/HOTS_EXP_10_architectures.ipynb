# On the full dataset

```python
sigma = None
pooling = False
homeinv = False
jitonic = [None,None] #[temporal, spatial]
jitter = False
R = 2
tau = 5
nbclust = [4,8,16]
filt = 2

timestr = '2021-03-29'
record_path = '../Records/EXP_03_NMNIST/'
dataset = 'nmnist'
nb_train = 60000
nb_test = 10000
ds = 10

nb_test = nb_test//ds
nb_train = nb_train//ds
print(f'training set size: {nb_train} - testing set: {nb_test}')
```

```python
kmax = 12
verbose=False
for nbclust in [[8,16],[4,8,16],[8,8,16],[16,8,16],[8,16,32]]:#,[16,32,64]]:
    print(f'architecture: {nbclust}')
    for name in ['homhots', 'hots']:
        hotshom, homeotest = netparam(name, filt, tau, nbclust, sigma, homeinv, jitter, timestr, dataset, R, subset_size = nb_train)
        hotshom.plotlayer()
        trainhistomap = hotshom.running(homeotest=homeotest, nb_digit = nb_train, outstyle='histo', subset_size = nb_train)
        testhistomap = hotshom.running(homeotest = homeotest, train=False, nb_digit=nb_test, jitonic=jitonic, subset_size = nb_test)
        print('---------Single histogram per record---------')
        kNN_score = knn(trainhistomap,testhistomap, k = kmax, weights = 'distance')
        print(f'Performances with  k-NN (k={kmax}): {np.round(kNN_score,3)*100}%')
        trainhistomapav = hotshom.running(homeotest=homeotest, nb_digit = nb_train, outstyle='histav', dataset=dataset, verbose=verbose)
        print('---------Averaged histograms-----------------')
        JS_score = knn(trainhistomapav,testhistomap, k = 1, weights = 'uniform', metric = JensenShannon)
        print(f'Performances with Jensen-Shannon distance: {np.round(JS_score,3)*100}%')
        EU_score = knn(trainhistomapav,testhistomap, k = 1, weights = 'uniform')
        print(f'Performances with Euclidian distance: {np.round(EU_score,3)*100}%')
```
